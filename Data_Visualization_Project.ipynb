```python
# -----------------------------
# Data Visualization Project - Tips Dataset
# Author: Mamuduru Sai Kathyayani
# -----------------------------

# Make plots appear inline
%matplotlib inline

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# -----------------------------
# Load Dataset
# -----------------------------
df = sns.load_dataset('tips')

# Show first 10 rows
df.head(10)

# Display dataset info
df.info()

# Statistical summary
df.describe()

# -----------------------------
# Line Plot: Trend of Total Bill by Day
# -----------------------------
plt.figure(figsize=(7,5))
sns.lineplot(data=df, x='day', y='total_bill', marker='o', color='blue', errorbar=None)
plt.title('Trend of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.savefig('line_plot_total_bill.png')  # Save plot
plt.show()

# Observation:
# Friday has the highest average total bill, Thursday the lowest.

# -----------------------------
# Bar Plot: Average Tip by Day and Gender
# -----------------------------
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='day', y='tip', hue='sex', palette='pastel', errorbar=None)
plt.title('Average Tip by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.legend(title='Sex')
plt.savefig('bar_plot_tip.png')
plt.show()

# Observation:
# Male customers tip slightly higher on average across most days.

# -----------------------------
# Scatter Plot: Tip vs Total Bill
# -----------------------------
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', style='time', s=100)
plt.title('Tip vs Total Bill by Sex and Time')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Sex & Time')
plt.savefig('scatter_plot.png')
plt.show()

# Observation:
# Tips increase with total bill. Dinner bills are generally higher than lunch bills.

# -----------------------------
# Boxplot: Outliers in Total Bill by Day
# -----------------------------
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='day', y='total_bill', hue='sex', palette='Set2')
plt.title('Outliers in Total Bill by Day and Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Sex')
plt.savefig('boxplot_total_bill.png')
plt.show()

# Observation:
# Saturday and Sunday have several high outliers; males tend to have higher bills.

# -----------------------------
# Save Dataset as CSV (Optional)
# -----------------------------
df.to_csv('tips_dataset.csv', index=False)

# -----------------------------
# Notebook Ready for GitHub
# - Displays dataset
# - Shows visualizations inline
# - Saves plots and dataset
```
